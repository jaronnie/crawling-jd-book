# 导入数据

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('store/jd-item-info-天文学-编程-thouhts-wenxue-生物.csv')
df.info()
```

### 删除重复的数据

```python
print(df.duplicated())
df = df.drop_duplicates()  
df
```

### 索引重排

```python
df = df.reset_index(drop=True)   
df
```

### 将publisher中的unkown和1和favorable_rate、comments_count中的unknown去除

```python
#将publisher中的unkown和1和favorable_rate、comments_count中的unknown去除
df = df.loc[(df.publisher!='unkown')&(df.publisher!='1')&(df.favorable_rate!='unknown')&(df.comments_count!='unknown'),:]   
df
```

### 将favorable_rate的数据类型从object转换float ，便于后期的求均值

```python
df.favorable_rate=df.favorable_rate.apply(lambda x:float(x))  
```

### 画图时显示中文

```python

plt.rcParams['font.sans-serif'] = [u'SimHei']
plt.rcParams['axes.unicode_minus'] = False  
```

### 将评论的万字换成0000

```python
#将评论的万字换成0000
df.comments_count=df.comments_count.apply(lambda x: str(x).replace('万','0000'))
df
```

### 将评论的+换成空字符

```python
#将评论的+换成空字符
df.comments_count=df.comments_count.apply(lambda x: int(str(x.replace('+',''))))
df
```

# 2.统计各个出版社的平均好评率

```python
f=[]
j=0
m=0
for i in df_publisher.values:
    #print(i)
    if(i>=10):     #取书的种类大于9的出版社，放入f中
        f.append(df_publisher.index[j])
        m=m+1
    j=j+1
f
```

```python
i=0
r=[]
for i in range(m):   #取书的种类大于9的出版社好评率的均值，放入r中
    r.append(df.loc[df.publisher==f[i],:].favorable_rate.mean()*100)
r
```

```python
plt.figure(figsize=(20,10))
plt.bar(f,r)
plt.title('各个出版社平均好评率')
plt.xlabel('出版社')
plt.ylabel('平均好评率')
plt.xticks(rotation=90)
plt.show()
```

```python
# #切词
import jieba
from collections import  Counter
from pyecharts.charts import WordCloud
import pyecharts.options as opts
name=df.loc[df.type=='编程','name']
name_str = ''
for s in name:
    name_str += s
name_str
result_cut = jieba.cut(name_str)

#容易被误分为词的符号与词
with open('data/language.txt','r',encoding='utf-8') as fp:
    data1 = fp.readlines()

add_list1 = []
for i in data1:
    add_list1.append(i.strip())
    
print(add_list1)

#统计词频
dic = {}
for value in result_cut:
    if value in add_list1:
        if value in dic:
            dic[value] += 1
        else:
            dic[value] = 1
        
print(dic)

#字典中以键值对为元组返回列表items()
item = dic.items()
items_list = list(item)

print(items_list)


(
    WordCloud()
    .add(series_name="热点分析", data_pair=items_list, word_size_range=[6, 66])
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="热点分析", title_textstyle_opts=opts.TextStyleOpts(font_size=23)
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
    )
    .render("basic_wordcloud.html")
)
```

```python
def word_cloud(df,type):
    name=df.loc[df.type==type,'name']
    name_str = ''
    for s in name:
        name_str += s
    result_cut = jieba.cut(name_str)

    #容易被误分为词的符号与词
    with open('data/hit_stopwords.txt','r',encoding='utf-8') as fp:
        data1 = fp.readlines()
    with open('data/cn_stopwords.txt','r',encoding='utf-8') as fp:
        data2 = fp.readlines()
    with open('data/scu_stopwords.txt','r',encoding='utf-8') as fp:
        data3 = fp.readlines()

    stop_list1 = []
    stop_list2 = []
    stop_list3 = []
    for i in data1:
        stop_list1.append(i.strip())
    for i in data2:
        stop_list2.append(i.strip())
    for i in data3:
        stop_list3.append(i.strip())

    #统计词频
    dic = {}
    for value in result_cut:
        if value in stop_list1:
            continue
        if value in stop_list2:
            continue
        if value in stop_list3:
            continue
        if value in dic:
            dic[value] += 1

        else:
            dic[value] = 1
    print(dic)
#     dic.pop('版')
#     dic.pop('博文')
    #字典中以键值对为元组返回列表items()
    item = dic.items()
    items_list = list(item)

    print(items_list)

    # # dict.pop("")
    (
        WordCloud()
        .add(series_name="热点分析", data_pair=items_list, word_size_range=[6, 66])
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title="热点分析", title_textstyle_opts=opts.TextStyleOpts(font_size=23)
            ),
            tooltip_opts=opts.TooltipOpts(is_show=True),
        )
        .render(type+"wordcloud.html")
    )
```

```python
word_cloud(df,'天文学')
word_cloud(df,'生物学')
word_cloud(df,'哲学')
```
